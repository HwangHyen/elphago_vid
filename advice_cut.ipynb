```python
count=0
for f in tqdm.tqdm(Path('data/raw/new').glob('*.png')):
    img = Image.open(f)
    np_img = np.array(img)
    Image.fromarray(np_img[700:770,490:790]).save(f'data/advice_cut/{count}.png')
    Image.fromarray(np_img[700:770,810:1110]).save(f'data/advice_cut/{count+1}.png')
    Image.fromarray(np_img[700:770,1130:1430]).save(f'data/advice_cut/{count+2}.png')
    count+=3
```
